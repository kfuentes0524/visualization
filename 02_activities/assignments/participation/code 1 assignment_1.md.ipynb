Making a basic figure with matplotlib
First, letâ€™s load our libraries

```python
# Import necessary libraries
import scipy
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import PIL
```

MATPLOTLIB
making our basic scatterplot

```python
# Set a random seed for reproducibility
np.random.seed(613) # Ensures the random numbers generated are the same every time

# Generate example data
x = np.arange(50) # Create an array with values from 0 to 49 (50 values in total)
y = np.random.randint(0,100,50) # Generate 50 random integers between 0 and 100

fig, ax = plt.subplots(figsize=(5,3)) # Set the figure size to 5x3 inches
ax.scatter(x,y)  # Plot points where 'x' is the horizontal axis and 'y' is the vertical axis
```

EXERCISE : BAR

```python
# Create a bar chart with the data
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y) # Plot vertical bars where 'x' is the category (horizontal axis) and 'y' is the height of each bar
```

EXERCISE : HISTOGRAM

```python
# Create a HISTOGRAM chart with the data
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

EXERCISE : PLOT

```python
# Create a PLOT chart with the data
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y) # Plot a line where 'x' represents the values on the horizontal axis, and 'y' represents the values on the vertical axis
```

LABEL AND TITLES
Adding axis labels

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

# Add a title and axis labels
ax.set_title("Total growth over time") # Set the title of the plot to describe the data being visualized
ax.set_ylabel("Total Growth") # Label the vertical axis to indicate the values represent growth
ax.set_xlabel("Years since start") # Label the horizontal axis to show it represents years since the start of tracking
```

Modifying labels and titles
dic : fontdic

```python
# Setting fonts
font1 = {"family": "sans-serif","color":"blue", "size":20} # Font settings for the title: sans-serif font, blue color, size 20
font2 = {"family": "monospace","color":"green", "size":14} # Font settings for the labels: monospace font, green color, size 14

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

# Add title and labels with customized fonts
ax.set_title("Total growth over time", fontdict = font1)  # Set the title with the font style defined in 'font1'
ax.set_ylabel("Total growth", fontdict = font2) # Label the vertical axis with the font style defined in 'font2'
ax.set_xlabel("Years since start", fontdict = font2) # Label the horizontal axis with the font style defined in 'font2'
```

```python
fig, ax = plt.subplots(figsize=(5,3))

# Plot a line graph with customized styles
ax.plot(x,y,
        marker = "*",   # Use a star ("*") marker to represent each data point
        color = "indigo",    # Set the color of the line to indigo
        linestyle = "--" ,    # Set the line style to dashed (denoted by "--")
        markeredgecolor = "indigo")    # Set the color of the marker edges to indigo (same as the line color)
```

Grid lines
.grid()

```python
fig, ax = plt.subplots(figsize=(5,3))

# Plot a line graph with customized styles
ax.plot(x,y,
        marker = "*",
        color = "indigo",
        linestyle = "--" ,
        markeredgecolor = "indigo")

# Add a grid to the plot for better visualization of the data points
ax.grid() # Enable the grid on the plot, making it easier to read the values and compare points
```

This plot shows how the scores of different groups of students (N1, N2, N3, N4, N5) changed from year to year. The x-axis represents the years, and the y-axis shows the scores. Each line represents a group of students, and the legend helps identify which line corresponds to which group.

```python
# Defining the 'years' as a tuple containing years from 2010 to 2020
years = ("2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020")

# Generating random data for different levels using numpy's random.rand() function
# Random numbers between 0 and 1 are multiplied to scale the values appropriately.
level1 = np.random.rand(11)*100  # Random values scaled to be between 0 and 100
level2 = np.random.rand(11)*200 + 100  # Random values scaled to be between 100 and 300
level3 = np.random.rand(11)*300 + 200  # Random values scaled to be between 300 and 500
level4 = np.random.rand(11)*400 + 300  # Random values scaled to be between 400 and 700
level5 = np.random.rand(11)*500 + 400  # Random values scaled to be between 500 and 900
```

```python
# Plotting the data for different levels over the years with labels for each level
plt.plot(years, level1, label = "N1")
plt.plot(years, level2, label = "N2")
plt.plot(years, level3, label = "N3")
plt.plot(years, level4, label = "N4")
plt.plot(years, level5, label = "N5")

plt.legend() # Adding a legend to the plot to differentiate between the different levels
plt.title("Certification exam") # Adding a title to the plot
plt.xlabel("years") # Adding labels to the x and y axes
plt.ylabel("score") # Adding labels to the x and y axes
plt.show() # Show the final plot with all customizations applied
```

Aesthetic Quality: The chart is simple yet effective. The default colors are pleasant, but they could be customized for better appeal.
Substantive Quality: The chart clearly shows how score amounts vary by years, answering the question effectively.
Perceptual Quality: The information is easy to interpret due to clear axes and well-defined labels.
