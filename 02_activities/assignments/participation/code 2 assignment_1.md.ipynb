```python
# letâ€™s load our libraries
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Setting a seed for reproducibility in random number generation
np.random.seed(613) # Ensures that the random numbers generated are the same every time
x = np.arange(50) # Creates an array of integers from 0 to 49
y1 = np.random.randint(0, 100,50) # Random integers between 0 and 100 for Person 1
y2 = np.random.randint(0, 100,50) # Random integers between 0 and 100 for Person 2

# Plotting data for Person 1
plt.plot (x,y1, label ="Person 1")  # Plots x against y1 and labels the line as "Person 1"
plt.show

# Plotting data for Person 2
plt.plot (x,y2, label = "Person 2")# Plots x against y1 and labels the line as "Person 2"
plt.legend()

# Adding a legend to the plot
frameon = True, #add frame around the legend
fontsize = 12, #change font size
ncol = 2, #specify number of columns
shadow = True 
plt.show

```

```python
# Adding a legend
# Add a â€˜labelâ€™ argument to each plot object

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right') # Call â€˜legendâ€™ and specify our legend location
fig.show()
```

Modifying our legend
By adding arguments to our legend object, we can change different elements of our legend

```python
# Change our legend location

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='upper left') # Call â€˜legendâ€™ and specify our legend location
fig.show()
```

```python
# Modify our legend location

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
frameon = True, #add frame around the legend
fontsize = 12, #change font size
ncol = 2, #specify number of columns
shadow = True )
fig.show()
```

```python
# Modify our legend location

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc="upper left",bbox_to_anchor =(1, 1))
fig.show()
```

Text and Annotations

```python
# Use the text() function to specify coordinates and text

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# Adding a text annotation to the plot at a specific position (x=40, y=50)
ax.text(40, 50, "This value is important!",     # The text to display at the coordinates (40, 50)
        ha = "center",                          # Horizontal alignment of the text (centered)
        color = "red",                          # Color of the text (red)
        size = 5)                               # Size of the text (5)
fig.show()
```

Positioning text annotations.
We can do this using transforms

```python
# Start by making a simple, empty set of axes
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

# This places the text ". Data:(1, 5)" at the data coordinates (1, 5), 
# where 'transform=ax.transData' ensures the text position is interpreted in data space.
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData )

# This places the text ". Axes:(0.5, 0.1)" at the coordinates (0.5, 0.1) relative to the axes. 
# The transform=ax.transAxes means the coordinates are in a normalized axes space (0 to 1 range).
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes )

# This places the text ". Figure:(0.2, 0.2)" at the coordinates (0.2, 0.2) relative to the entire figure.
# The transform=fig.transFigure ensures the coordinates are in figure space (from 0 to 1, relative to the figure size).
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

```python
# Annotating with arrows

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# # Add annotation with an arrow to the plot
ax.annotate("This is important",
xy=(10, 95), xytext=(15, 99), # The arrow is styled as a "wedge" and colored black, 
arrowprops = dict(arrowstyle = "wedge",
color = "Black"))
fig.show()
```

Removing tick marks and labels.
We can remove tick marks or labels from our axis like so:
Use locator to remove ticks and labels
Use formatter to remove only labels

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.NullLocator())# Locator to remove ticks and label
ax.yaxis.set_major_formatter(plt.NullFormatter()) # formattes to remove only label
```

```python
# Rotating axis labels

plt.style.use("_mpl-gallery-nogrid")
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")
plt.xticks(rotation=25, ha="right")  # Rotate the x-axis tick labels by 25 degrees and align them to the right
plt.title("Number of person")
plt.xlabel("years")
plt.ylabel("Person")
```

SUBPLOT.
Now we want to have two plots next to each other, so we just have to define multiple axes and their relative positions.

```python
# # letâ€™s load our libraries
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# Setting a seed for reproducibility in random number generation 
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy"
,
"Zoro"
,
"Nami"
,
"Usopp"
,
"Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# Create a figure with two subplots arranged in one row and two columns, with a specified size
fig,(ax1, ax2) = plt.subplots(ncols=2,nrows=1,figsize=(7, 3))

ax1.scatter (x1,y1, color = "red") # Scatter plot on the first subplot (ax1) with red color
ax2.bar(x2,y2, color = "green") # Bar plot on the second subplot (ax2) with green color
```

```python
# Modify figure layout
# Layouts Comparison

plt.style.use("_classic_test_patch")

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],['ax2', 'ax3']],figsize=(7, 4), layout = "constrained")

someaxes["ax1"].scatter(x1,y1, color = "red")
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```
